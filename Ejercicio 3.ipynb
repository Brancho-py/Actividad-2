```python
from enum import Enum


class tipoCom(Enum):
    GASOLINA = "GASOLINA"
    BIOETANOL = "BIOETANOL"
    DIESEL = "DIESEL"
    BIODISESEL = "BIODISESEL"
    GAS_NATURAL = "GAS_NATURAL"

class tipoA(Enum):
    CIUDAD = "CIUDAD"
    SUBCOMPACTO = "SUBCOMPACTO"
    COMPACTO = "COMPACTO"
    FAMILIAR = "FAMILIAR"
    EJECUTIVO = "EJECUTIVO"
    SUV = "SUV"

class tipoColor(Enum):
    BLANCO = "BLANCO"
    NEGRO = "NEGRO"
    ROJO = "ROJO"
    NARANJA = "NARANJA"
    AMARILLO = "AMARILLO"
    VERDE = "VERDE"
    AZUL = "AZUL"
    VIOLETA = "VIOLETA"


class Automovil:
    def __init__(self, marca, modelo, motor, tipo_combustible,
                 tipo_automovil, numero_puertas, cantidad_asientos,
                 velocidad_maxima, color):
        self.marca = marca
        self.modelo = modelo
        self.motor = motor
        self.tipoCombustible = tipo_combustible
        self.tipoAutomovil = tipo_automovil
        self.numeroPuertas = numero_puertas
        self.cantidadAsientos = cantidad_asientos
        self.velocidadMaxima = velocidad_maxima
        self.color = color
        self.velocidadActual = 0


    def setVelocidadActual(self, velocidad):
        self.velocidadActual = velocidad

    def getVelocidadActual(self):
        return self.velocidadActual

    def acelerar(self, incrementoVelocidad):
        if (self.velocidadActual + incrementoVelocidad) < self.velocidadMaxima:
            self.velocidadActual = self.velocidadActual + incrementoVelocidad
        else:
            print("No se puede incrementar a una velocidad superior a la máxima.")

    def desacelerar(self, decrementoVelocidad):
        if (self.velocidadActual - decrementoVelocidad) > 0:
            self.velocidadActual = self.velocidadActual - decrementoVelocidad
        else:
            print("No se puede decrementar a una velocidad negativa.")

    def frenar(self):
        self.velocidadActual = 0

    def imprimir(self):
        print(f"\n--- DATOS: {self.marca} ---")
        print(f"Modelo: {self.modelo} | Motor: {self.motor}")
        print(f"Combustible: {self.tipoCombustible.name} | Tipo: {self.tipoAutomovil.name}")
        print(f"Velocidad Máxima: {self.velocidadMaxima} | Color: {self.color.name}")


def seleccionar_opcion(clase_enum, nombre):
    opciones = list(clase_enum)
    print(f"\nSeleccione {nombre}:")
    for i, op in enumerate(opciones, 1):
        print(f"{i}. {op.name}")
    while True:
        try:
            res = int(input("Elija el número: "))
            if 1 <= res <= len(opciones):
                return opciones[res-1]
        except ValueError: pass
        print("Opción inválida.")


autos = []
n = int(input("¿Cuántos automóviles desea registrar?: "))

for i in range(n):
    print(f"\n>>> REGISTRO AUTO #{i+1} <<<")
    marca = input("Marca: ")
    modelo = int(input("Modelo (Año): "))
    motor = int(input("Motor (Cilindrada): "))

    t_comb = seleccionar_opcion(tipoCom, "Tipo de Combustible")
    t_auto = seleccionar_opcion(tipoA, "Tipo de Automóvil")
    t_col = seleccionar_opcion(tipoColor, "Color")

    puertas = int(input("Número de puertas: "))
    asientos = int(input("Cantidad de asientos: "))
    vmax = int(input("Velocidad máxima: "))

    nuevo_auto = Automovil(marca, modelo, motor, t_comb, t_auto, puertas, asientos, vmax, t_col)
    autos.append(nuevo_auto)


print("\n" + "="*30)
print("EJECUTANDO PRUEBAS DE VELOCIDAD")
print("="*30)

for a in autos:
    a.imprimir()


    a.setVelocidadActual(100)
    print(f"Velocidad actual establecida en: {a.getVelocidadActual()}")


    print(f"Acelerando +20...")
    a.acelerar(20)
    print(f"Velocidad actual: {a.getVelocidadActual()}")


    print(f"Desacelerando -50...")
    a.desacelerar(50)
    print(f"Velocidad actual: {a.getVelocidadActual()}")


    print(f"Frenando...")
    a.frenar()
    print(f"Velocidad actual: {a.getVelocidadActual()}")


    print(f"Intentando desacelerar -20 estando en 0:")
    a.desacelerar(20)
```
